## Cálculo da Produção de Milho por km² (em milhões de R$)

A fórmula utilizada para calcular a produção de milho por km² em milhões de reais é:

$$
\text{Produção de Milho(R\$) por km² no ano X} = \left( \frac{\text{Produção de Milho (R\$) no ano X}}{\text{Área total (km²)}} \right) 
$$

Onde:

- **Produção de Milho (R$) no ano X**: Valor da produção de milho em reais para um determinado ano.
- **Área total (km²)**: Área total considerada no cálculo (**1.247.954,666 km²**).



```python
para_km = 1247954.666

colunas_producao = df_para_milho_rs.columns[3:]
for i , linha in df_para_milho_rs.iterrows():
    if i >= 3:
        producao_milho_para_por_km2 = (linha[colunas_producao]/para_km)
```

```python
df_para_milho_rs.head()
```

```python
# Inicializando uma lista vazia para armazenar os valores do dicionário
dados = []

# Preenchendo os dados com um for
for valor in producao_milho_para_por_km2:
    dados.append({'Produção Milho (R$)/km²': valor})

# Criando o DataFrame com os dados preenchidos
df_producao_milho_km = pd.DataFrame(dados)

# Exibindo o DataFrame invertido
print(df_producao_milho_km.head(22))
```

```python
for i, row in df_producao_milho_km.iterrows():
    print(f"{i}) {row}")
```

```python
# Inicializando a lista para armazenar os valores a partir da posição 12
valores_filtrados = []

# Percorrendo o DataFrame a partir da linha 12 até o final
for index, row in df_producao_milho_km.iloc[12:].iterrows():
    valores_filtrados.append(row['Produção Milho (R$)/km²'])

# Exibindo a lista de valores filtrados
print(valores_filtrados)

print(len(valores_filtrados))
```

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df_desmatamento = pd.read_csv("PRODES(DESMATAMENTO)/terrabrasilis_legal_amazon_02_02_2025_taxas_de_desmatamento_estadual_por_ano.csv", delimiter=";", engine="python")
```

```python
df_desmatamento.head()
```

```python
df_desmatamento.dtypes
```

```python
#Substituir vírgula por ponto na coluna "area km²" e converter para float
df_desmatamento["area km²"] = df_desmatamento["area km²"].str.replace(".", "", regex=False)  # Remove o ponto separador de milhar
df_desmatamento["area km²"] = df_desmatamento["area km²"].str.replace(",", ".", regex=False).astype(float)
```

```python
df_desmatamento.dtypes
```

```python
df_desmatamento["area km²"].head()
```

```python
menor_ano = df_desmatamento["year"].min()
maior_ano = df_desmatamento["year"].max()

print(f"Menor ano: {menor_ano}")
print(f"Maior ano: {maior_ano}")
```

```python
def acumulador_area_desmatada():
    # Filtra os dados dentro do intervalo desejado
    df_filtrado = df_desmatamento[(df_desmatamento["year"] >= 1999) & (df_desmatamento["year"] <= 2020)]
    
    # Soma a área total desmatada no período
    total_desmatado = df_filtrado["area km²"].sum()
    
    return total_desmatado

# Chamando a função e imprimindo o resultado
print(f"Total de área desmatada (1999-2020): {acumulador_area_desmatada()} km²")
```

```python
def acumulador_para_desmatado():
    # Certifique-se de que a coluna "year" está no formato numérico
    df_desmatamento["year"] = df_desmatamento["year"].astype(int)

    # Filtra os dados para o intervalo de 1999 a 2020 e apenas para o estado do Pará
    df_filtrado = df_desmatamento[
        (df_desmatamento["year"] >= 1999) & 
        (df_desmatamento["year"] <= 2020) & 
        (df_desmatamento["uf"] == "Pará")  # Filtra apenas o estado do Pará
    ]

    # Soma a área total desmatada no Pará no período
    total_desmatado_para = df_filtrado["area km²"].sum()

    return total_desmatado_para

# Chamando a função e imprimindo o resultado
print(f"Total de área desmatada no Pará (1999-2020): {acumulador_para_desmatado()} km²")
```

```python
# Obtendo os valores das funções
total_brasil = acumulador_area_desmatada()
total_para = acumulador_para_desmatado()

# Criando o histograma
plt.figure(figsize=(8, 5))
plt.bar(["Amazônia Legal", "Pará"], [total_brasil, total_para], color=["blue", "green"])

# Adicionando rótulos e título
plt.xlabel("Território")
plt.ylabel("Área Desmatada (km²)")
plt.title("Comparação da Área Desmatada (1999-2020)")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Exibir os valores exatos em cada barra
for i, valor in enumerate([total_brasil, total_para]):
    plt.text(i, valor + 1000, f"{valor:.2f}", ha="center", fontsize=12, fontweight="bold")

plt.show()
```

```python
total_brasil = acumulador_area_desmatada()  # Total de área desmatada no Brasil
total_para = acumulador_para_desmatado()  # Total de área desmatada no Pará

#https://terrabrasilis.dpi.inpe.br/app/dashboard/deforestation/biomes/legal_amazon/rates : Taxas de desmatamento - Amazônia Legal - Estados

# Calculando a porcentagem
porcentagem_para = (total_para / total_brasil) * 100

print(f"A área desmatada no Pará representa {porcentagem_para:.2f}% do total de área desmatada no Brasil.")
```
