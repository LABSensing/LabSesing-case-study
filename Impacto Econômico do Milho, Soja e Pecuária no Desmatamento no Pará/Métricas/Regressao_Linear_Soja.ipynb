# Regressão Linear:

## Aplicado com a densidade: Produção de Soja(R$)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df_dados_geral= pd.read_excel("../Métricas/Aqui/Produção de Milho(R$)km + Produção Rebanho Bovino(R$)km + Produção de Soja(R$) + desmatamento.xlsx")
df_dados_geral.head(0)
```

```python
df_dados_geral = df_dados_geral.drop(columns=["uf"], errors="ignore")
df_dados_geral = df_dados_geral.rename(columns={"Produção Rebanho Bovino(R$)/km²": "Quantitativo Bovino/km²"})
df_dados_geral = df_dados_geral.rename(columns={"area km²": "Area Desmatada km²"})

# Exibir o cabeçalho do DataFrame
df_dados_geral.head()
```

```python
df_dados_geral = df_dados_geral.rename(columns={"year": "Ano"})

# Exibir o cabeçalho do DataFrame
df_dados_geral.head()
```

```python
df_dados_geral.head()
```

```python
df_dados_geral.dtypes
```

```python
X = df_dados_geral[["Area Desmatada km²"]]
y = df_dados_geral[["Produção Soja(R$)/km²"]]  
```

```python
print(X.shape)
print(y.shape)
```

```python
modelo = LinearRegression()  # Cria o modelo
modelo.fit(X, y)  # Treina o modelo
```

```python
# Verificando o shape de X e y
print(f"Shape de X: {X.shape}")
print(f"Shape de y: {y.shape}")
```

```python
y_pred = modelo.predict(X)  # Previsões para os dados de entrada
```

```python
plt.scatter(X, y, color='blue', label='Dados observados')  # Dados reais
plt.plot(X, y_pred, color='red', label='Regressão linear')  # Linha de regressão

plt.xlabel("Área Desmatada (km²)", fontsize=20)
plt.ylabel("Produção Soja (R$)/km²", fontsize=20)
plt.title("Regressão Linear: Área Desmatada vs. Produção Soja (R$)/km²", fontsize=20)
plt.legend(fontsize=15)

plt.show()
```

```python
r2 = modelo.score(X, y)
print(f"Coeficiente de determinação (R²): {r2:.2f}")
```

```python
r2_percentual = r2 * 100  # Convertendo R² para porcentagem
print(f"Coeficiente de determinação (R²): {r2_percentual:.2f}%")
```
