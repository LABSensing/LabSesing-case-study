## Aplicado com a densidade: Quantitativo Bovino(km)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df_dados_geral= pd.read_excel("../Métricas/Aqui/Produção de Milho(R$)km + Produção Rebanho Bovino(R$)km + Produção de Soja(R$) + desmatamento.xlsx")
df_dados_geral.head()
```

```python
df_dados_geral = df_dados_geral.rename(columns={"Produção Rebanho Bovino(R$)/km²": "Quantitativo Bovino/km²"})
```

```python
df_dados_geral.head()
```

```python
X = df_dados_geral[["area km²"]]  # Variável independente (área desmatada)
y = df_dados_geral[["Quantitativo Bovino/km²"]]  # Variável dependente (quantitativo bovino)
```

```python
print(X.shape)
print(y.shape)
```

```python
modelo = LinearRegression()  # Cria o modelo
modelo.fit(X, y)  # Treina o modelo
```

```python
y_pred = modelo.predict(X)  # Previsões para os dados de entrada
```

```python
plt.scatter(X, y, color='blue', label='Dados observados')  # Dados reais
plt.plot(X, y_pred, color='red', label='Regressão linear')  # Linha de regressão
plt.xlabel("Área Desmatada (km²)", fontsize=20)
plt.ylabel("Quantitativo Bovino(km²)", fontsize=20)
plt.title("Regressão Linear: Área Desmatada vs. Quantitativo Bovino/km²", fontsize=20)
plt.legend(fontsize=15)
plt.show()
```

```python
r2 = modelo.score(X, y)
print(f"Coeficiente de determinação (R²): {r2:.2f}")
```

```python
r2_percentual = r2 * 100  # Convertendo R² para porcentagem
print(f"Coeficiente de determinação (R²): {r2_percentual:.2f}%")
```

```python
df_gado = pd.read_excel("Rebanho-bovino/Cabeca-Gado(R$).xls")
df_gado.head(100)
```

```python
df_gado_para = df_gado[df_gado["State"] == "Pará"]
```

```python
df_gado_para.head()
```

```python
# Inicializando uma lista vazia para armazenar os valores do dicionário
dados = []

# Selecionando as colunas de produção (de 1999 até 2020)
colunas_producao = df_gado_para.columns[3:] 
for index, row in df_gado_para.iterrows():
    for valor in colunas_producao:
        dados.append({'Quantitativo Bovino': row[valor]})
df_num_gado = pd.DataFrame(dados)
print(df_num_gado.head(22))
```

```python
df_num_gado.head()
```

```python
df_dados_geral["area km²"].head()
```

```python
X = df_dados_geral[["area km²"]]
y= df_num_gado
```

```python
print("Formato de X:", X.shape)
print("Formato de y:", y.shape)
```

```python
modelo = LinearRegression()  # Cria o modelo
modelo.fit(X, y)  # Treina o modelo
```

```python
y_pred = modelo.predict(X)  # Previsões para os dados de entrada
```

```python
plt.scatter(X, y, color='blue', label='Dados observados')  # Dados reais
plt.plot(X, y_pred, color='red', label='Regressão linear')  # Linha de regressão
plt.xlabel("Área Desmatada (km²)")
plt.ylabel("Quantitativo Bovino")
plt.title("Regressão Linear: Área Desmatada vs. Quantitativo Bovino")
plt.legend()
plt.show()
```

```python
r2 = modelo.score(X, y)
print(f"Coeficiente de determinação (R²): {r2:.2f}")
```
