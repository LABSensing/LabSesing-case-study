## Necessário criar listas com os valores anuais para cada função:

```python
import matplotlib.pyplot as plt

# Lista de anos (colunas do DataFrame)
anos = [str(ano) for ano in range(1999, 2021)]  # 1999 a 2020

# Dados do Pará
producao_para = []
for ano in anos:
    try:
        # Chama cada função acumulador_gado_para_XXXX()
        valor = eval(f"acumulador_gado_para_{ano}()")  # Nota: eval pode ser inseguro - veja alternativa abaixo
        producao_para.append(valor)
    except:
        producao_para.append(0)  # Caso a função/ano não exista

# Dados da Amazônia
producao_amazonia = []
for ano in anos:
    try:
        # Chama cada função acumulador_gado_amazonia_{ano}()
        valor = eval(f"acumulador_gado_amazonia_{ano}()")
        producao_amazonia.append(valor)
    except:
        producao_amazonia.append(0)
```

```python
# Geração da lista de anos (como strings)
anos = [str(ano) for ano in range(1999, 2021)]  # 1999 a 2020 (inclusive)

# Verificação (para provar que funciona)
print("Anos gerados:", anos)
print("Total de anos:", len(anos))
print("Primeiro ano:", anos[0])
print("Último ano:", anos[-1])
```

##  Criação do gráfico 

```python
import matplotlib.pyplot as plt
import numpy as np

# Configuração do gráfico
plt.figure(figsize=(15, 8))
bar_width = 0.35
indices = np.arange(len(anos))

# Barras para o Pará (azul)
plt.bar(
    indices - bar_width/2,
    producao_para,
    width=bar_width,
    label='Pará',
    color='#1f77b4',
    edgecolor='black',
    linewidth=0.5
)

# Barras para a Amazônia (laranja)
plt.bar(
    indices + bar_width/2,
    producao_amazonia,
    width=bar_width,
    label='Amazônia',
    color='#ff7f0e',
    edgecolor='black',
    linewidth=0.5
)

# Formatação aprimorada
plt.title('Quantitativo Bovino (1999-2020)\nPará vs. Região Amazônica', 
          fontsize=16, pad=20, weight='bold')
plt.xlabel('Ano', fontsize=12, labelpad=10)
plt.ylabel('Quantitativo Bovino', fontsize=12, labelpad=10)

# Formatação do eixo Y com "Unidade(s)"
plt.gca().yaxis.set_major_formatter(lambda x, pos: f'{int(x)} Unidade{"s" if x != 1 else ""}')

# Ajustes nos eixos
plt.xticks(indices, anos, rotation=45, ha='right')
plt.yticks(fontsize=10)

# Grade e legenda
plt.grid(axis='y', linestyle='-', alpha=0.6, color='lightgray')
plt.legend(fontsize=12, framealpha=1, facecolor='white')

# Adicionando valores nas barras (com "Unidade(s)")
for i in indices:
    plt.text(
        i - bar_width/2, 
        producao_para[i], 
        #f'{int(producao_para[i])} Unidade{"s" if producao_para[i] != 1 else ""}',
        ha='center', 
        va='bottom', 
        fontsize=9,
        weight='bold',
        bbox=dict(facecolor='#1f77b4', alpha=0.8, boxstyle='round,pad=0.3')
    )
    plt.text(
        i + bar_width/2, 
        producao_amazonia[i], 
        #f'{int(producao_amazonia[i])} Unidade{"s" if producao_amazonia[i] != 1 else ""}',
        ha='center', 
        va='bottom',
        fontsize  =9,
        weight='bold',
        bbox=dict(facecolor='#ff7f0e', alpha=0.8, boxstyle='round,pad=0.3')
    )

# Ajustes finais
plt.tight_layout()
plt.show()
```
