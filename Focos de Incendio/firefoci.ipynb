```python
plt.figure(figsize=(18, 10))

for i, estado in enumerate(estados):
    # Use the estado variable to access the correct column
    subset = merged_data[merged_data[estado] > 0]
    dados_estado = subset[estado].values  # Get the data for the current state

    anos_str = merged_data['Anos'].astype(str).tolist() # Convert Anos column to strings
    plt.plot(anos_str, dados_estado, marker='o', linestyle='-', label=estado)

plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of fireoutbreaks", fontsize=14)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()  # Add a legend to the plot
plt.show()  # Display the plot
```
