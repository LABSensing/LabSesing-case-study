#### Importando Pandas

```python
import pandas as pd
```

```python
df_dados_soja_br= pd.read_excel("./Dados_Brasil/Produção-de-Soja-Brasil.xls")
df_dados_cabecas_br= pd.read_excel("./Dados_Brasil/Efetivo-Bovino-Brasil.xls")
df_dados_milho_br= pd.read_excel("./Dados_Brasil/Produção-de-Milho-Brasil.xls")
```
