#### Pr√© processamento

```python
import pandas as pd
import os

# Configurar Pandas para exibir n√∫meros inteiros completos
pd.options.display.float_format = '{:.00f}'.format
try:
    df_area = pd.read_csv("IBGE_area_geografica_98-22.csv", encoding="utf-8", sep=";", skiprows=1)
    print(f"  ‚úî IBGE_area_geografica_98-22.csv carregado como df_area \n")
    print(df_area.dtypes)
except Exception as e:
        print(f"Erro ao ler arquivo : {e}")


#filtrando estados de interesse
estados = ["Acre", "Amazonas", "Amap√°", "Maranh√£o", "Par√°", "Roraima", "Tocantins", "Rond√¥nia", "Mato Grosso"]
df_area = df_area.query("@estados in Estado")

print(f"\n  üåé Filtrados apenas os estados desejados em df_area")

#removendo colunas indesejadas
col_para_remover = [col for col in ["Codigo", "C√≥digo", "Sigla"] if col in df_area.columns]
df_area = df_area.drop(columns = col_para_remover)

print(f"  üóëÔ∏è Colunas removidas de df_area: {col_para_remover}")

#convertendo valores str para int
df_area.iloc[:, 1:] = df_area.iloc[:, 1:].replace(",", ".", regex=True)  # Troca v√≠rgula por ponto
for col in df_area.iloc[:, 1:].columns:
    df_area[f"{col}"] = pd.to_numeric(df_area[f"{col}"], errors='coerce')

print(f"  üîÑ Colunas 1998 a 2022 convertidas para tipo inteiro")
```

#### Calculando M√©dia

```python
df_area["M√©dia Territorial"] = df_area.iloc[:, 1:].mean(axis=1)
```
