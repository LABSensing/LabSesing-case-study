```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df_desmatamento_csv = pd.read_csv("PRODES(DESMATAMENTO)/terrabrasilis_legal_amazon_02_02_2025_taxas_de_desmatamento_estadual_por_ano.csv",sep=";")
df_desmatamento_csv.head(20)
```

```python
df_desmatamento_csv.dtypes
```

```python
# 1. Remover espaços em branco extras
df_desmatamento_csv['uf'] = df_desmatamento_csv['uf'].str.strip()
df_desmatamento_csv.head()
```

```python
df_desmatamento_csv['area km²'] = df_desmatamento_csv['area km²'].replace({'\.': '', ',': '.'}, regex=True)
df_desmatamento_csv['area km²'] = pd.to_numeric(df_desmatamento_csv['area km²'], errors='coerce')
```

```python
df_desmatamento_csv.dtypes
```

```python
df_desmatamento_csv['area km²'].head(20)
```

```python
# Filtrar para o estado do Pará
df_para = df_desmatamento_csv[df_desmatamento_csv['uf'] == 'Pará']
```

```python
df_para.head(21)
```

```python
df_para.dtypes
```

```python
# Exportar o DataFrame filtrado para um arquivo Excel
df_para.to_excel("desmatamento_1999_2020_Para.xlsx", index=False)

print("\nArquivo 'desmatamento_1999_2020_Para.xlsx' foi gerado com sucesso.")
```
