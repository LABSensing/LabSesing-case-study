```python
df_milho_rs = pd.read_excel("Producao-Milhos/Produção - milho(R$).xls")
df_milho_rs.head(27)
```

```python
# Filtrar apenas as linhas onde o Estado é "Pará"
df_para_milho_rs = df_milho_rs[df_milho_rs["Estado"] == "Pará"]
df_para_milho_rs.head(1)
```

## Cálculo da Produção de Milho por km² (em milhões de R$)

A fórmula utilizada para calcular a produção de milho por km² em milhões de reais é:

$$
\text{Produção de Milho por km² (milhões de R\$) no ano X} = \left( \frac{\text{Produção de Milho (R\$) no ano X}}{\text{Área total (km²)}} \right) \times 10^6
$$

Onde:

- **Produção de Milho (R$) no ano X**: Valor da produção de milho em reais para um determinado ano.
- **Área total (km²)**: Área total considerada no cálculo (**1.247.954,666 km²**).
- **10⁶**: Fator de conversão para expressar o valor em **milhões de reais**.


```python
#https://www.agenciapara.com.br/pagina/47/para
# Definindo o valor da área (em km²)
para_km = 1247954.666
colunas_producao = df_para_milho_rs.columns[3:]
for i , linha in df_para_milho_rs.iterrows():
    if i >= 3:
        producao_milho_para_por_km2 = (linha[colunas_producao]/para_km)*1000000

```

```python
# Definindo o valor da área (em km²)
para_km = 1247954.666

# Selecionando as colunas de produção
colunas_producao = df_para_milho_rs.columns[3:]

# Iterando pelas linhas e modificando os valores diretamente
for i, linha in df_para_milho_rs.iterrows():
    if i >= 3:
        # Modificando diretamente as colunas de produção
        df_para_milho_rs.loc[i, colunas_producao] = (linha[colunas_producao] / para_km) * 1000000

# Verificando os primeiros valores após a modificação
print(df_para_milho_rs.head())


```

```python
# Inicializando uma lista vazia para armazenar os valores do dicionário
dados = []

# Preenchendo os dados com um for
for valor in producao_milho_para_por_km2:
    dados.append({'Produção Milho (R$)/km²': valor})

# Criando o DataFrame com os dados preenchidos
df_producao_milho_km = pd.DataFrame(dados)

# Exibindo o DataFrame invertido
print(df_producao_milho_km.head(22))
```

```python
df_desmatamento = pd.read_excel("PRODES(DESMATAMENTO)/desmatamento_1999_2020_PARA.xlsx")


# Exibir as primeiras linhas do DataFrame filtrado e ordenado
print(df_desmatamento.head(22))
```

```python
df_desmatamento['area km²'] = df_desmatamento['area km²'].astype(float)
```

```python
df_desmatamento.head(20)
```

```python
# Criar uma cópia do DataFrame original
df_desmatamento_crescente = df_desmatamento.copy()

# Filtrar as linhas onde o ano é maior ou igual a 1990
df_desmatamento_crescente = df_desmatamento_crescente[df_desmatamento_crescente['year'] >= 1990]

# Ordenar o DataFrame pela coluna 'year' em ordem crescente
df_desmatamento_crescente = df_desmatamento_crescente.sort_values(by='year')

# Resetar os índices para começar do 0
df_desmatamento_crescente = df_desmatamento_crescente.reset_index(drop=True)

# Exibir o DataFrame final
print(df_desmatamento_crescente)
```

```python
df_desmatamento_crescente.head(22)
```

```python
df_desmatamento_crescente.dtypes
```

```python
# Concatenar os dois DataFrames por linhas (verticalmente), ignorando os índices
df_producao_milho_rs_para_km = pd.concat([df_desmatamento_crescente, df_producao_milho_km], axis=1)

# Exibir o DataFrame resultante
df_producao_milho_rs_para_km.head()
```

```python
# Exportar o DataFrame filtrado para um arquivo Excel
df_producao_milho_rs_para_km.to_excel("Produção de Milho(R$)-km + desmatamento.xlsx", index=False)

print("\nArquivo 'Produção de Milho(R$)-km + desmatamento' foi gerado com sucesso.")
```
